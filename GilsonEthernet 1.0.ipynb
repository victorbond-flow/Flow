```python
%load_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
from gilson_ethernet2 import GilsonSession
from rack2 import Rack, Rackcommands

# --- Physical setup info ---
n_cols = 6
n_rows = 16
rack_home_x = 3.8
rack_home_y = 2.3
X_vial1 = 35.5
Y_vial1 = 7.2       # Y for centre of vial 1 after calibration
offset_x = X_vial1 
offset_y = Y_vial1
vial2vial_x = 16.54       # mm between adjacent columns
vial2vial_y = 17.77       # mm between adjacent rows
groundlevel_height = 0.0 # mm; tray surface (Z=0 reference for now)

# --- Default vial parameters (for every vial in this rack) ---
vial_volume_max = 2.0        # mL
vial_usedvolume_max = 1.8    # mL          NOTE: ALL VALUES ARE CURRENTLY PLACEHOLDERS - NEED TO UPDATE
vial_height = 25.0           # mm                 ONCE MEASUREMENTS HAVE BEEN MADE
vial_free_depth = 2.0        #mm           

# --- Create the rack ---
rack = Rack(
    array_dimensions=(n_cols, n_rows),
    offset_x=offset_x,
    offset_y=offset_y,
    vial2vial_x=vial2vial_x,
    vial2vial_y=vial2vial_y,
    groundlevel_height=groundlevel_height,
    staggered=True
    vial_volume_max=vial_volume_max,
    vial_usedvolume_max=vial_usedvolume_max,
    vial_height=vial_height,
    vial_free_depth=vial_free_depth
)

# --- Connect Gilson + create rack commands ---
g = GilsonSession("192.168.1.101")

rack_cmds = Rackcommands(gilson_session=g, rack=rack)
rack_cmds.rack_home_x = rack_home_x
rack_cmds.rack_home_y = rack_home_y

# --- Z safety limits ---
g.Z_SAFE = 45          # Lower Z limit (mm)
g.Z_MAX_SAFE = 120     # Upper Z limit (mm)
g.Z_WORKING_MIN = 0    # Lowest Z position allowed during vial operations (mm)
```
