From Gilson Ethernet Utility

Current IP is 192.168.1.101, Port is 50190, Unit ID is 35

```python
def port_scan(ip, start=5000, end=50200):
    open_ports = []
    for port in range(start, end):
        with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
            s.settimeout(0.2)
            if s.connect_ex((ip, port)) == 0:
                open_ports.append(port)
    return open_ports

open_ports = port_scan('192.168.1.101')
print("Open ports detected:", open_ports)
```

Apparently the only open port is 50185

```python
# Gilson autosampler IP and port
AUTOSAMPLER_IP = '192.168.1.101'
ADMIN_PORT = 50185
```

```python
def test_connection(ip, port, timeout=3):
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.settimeout(timeout)
        try:
            s.connect((ip, port))
            print(f"Successfully connected to {ip}:{port}")
        except Exception as e:
            print(f"Connection error: {e}")

test_connection(AUTOSAMPLER_IP, ADMIN_PORT)
```

```python
def request_instrument_port(ip, admin_port):
    request_xml = "<EthernetMessage><RequestPort/></EthernetMessage>"
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.settimeout(5)
        try:
            s.connect((ip, admin_port))
            print(f"Connected to Admin port {admin_port}")
            s.sendall(request_xml.encode('ascii'))
            response = s.recv(4096).decode('ascii')
            print("Response received:", response)
            # Parse assigned port number from XML
            assigned_port = int(response.split('>')[2].split('<')[0])
            print(f"Assigned Communication Port: {assigned_port}")
            return assigned_port
        except Exception as e:
            print(f"Error during port request: {e}")
            return None

# Run to get the actual communication port
assigned_port = request_instrument_port(AUTOSAMPLER_IP, ADMIN_PORT)
```

```python
def direct_identify(ip, port):
    identify_xml = "<EthernetMessage><Command Name='Identify'/></EthernetMessage>"
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.settimeout(5)
        try:
            s.connect((ip, port))
            print(f"Connected directly to {ip}:{port}, sending 'Identify'")
            s.sendall(identify_xml.encode('ascii'))
            response = s.recv(4096).decode('ascii')
            print(f"Instrument 'Identify' response:\n{response}")
        except Exception as e:
            print(f"Error during Identify command: {e}")

# Run the Identify command directly
direct_identify(AUTOSAMPLER_IP, ADMIN_PORT)
```

```python
def send_command(ip, port, command_name):
    command_xml = f"<EthernetMessage><Command Name='{command_name}'/></EthernetMessage>"
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.settimeout(5)
        try:
            s.connect((ip, port))
            s.sendall(command_xml.encode('ascii'))
            response = s.recv(4096).decode('ascii')
            print(f"Instrument response to '{command_name}': {response}")
        except Exception as e:
            print(f"Error sending command '{command_name}': {e}")

# Example: Identify command
if assigned_port:
    send_command(AUTOSAMPLER_IP, assigned_port, 'Identify')
```

```python
def request_admin_privileges(ip, port):
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.settimeout(3)
        try:
            s.connect((ip, port))
            admin_request = "<EthernetMessage><Command Name='Admin'/></EthernetMessage>"
            s.sendall(admin_request.encode('ascii'))
            response = s.recv(1024).decode('ascii')
            print("Admin privileges response:", response)
            return "Success" in response
        except Exception as e:
            print(f"Error requesting admin privileges: {e}")
            return False

# Request admin privileges using the received port
if instrument_port:
    admin_granted = request_admin_privileges(AUTOSAMPLER_IP, instrument_port)
    print("Admin privileges granted:", admin_granted)
```
