```python
import ftd2xx
import time

class Gilson4020PumpUSB:
    def __init__(self, ftdi_serial=b'G0SX4P41'):
        self.device = ftd2xx.openEx(ftdi_serial, 1)  # OPEN_BY_SERIAL_NUMBER = 1
        self.device.setTimeouts(1000, 1000)

    def send_command(self, command):
        self.device.write((command + "\r").encode('ascii'))
        time.sleep(0.2)
        response = self.device.read(1024)
        return response.decode('ascii', errors='ignore').strip()

    def identify(self):
        return self.send_command("Identify")

    def home(self):
        return self.send_command("Home")

    def close(self):
        self.device.close()
```

```python
pump = Gilson4020PumpUSB()
print("Home:", pump.home())
print("Serial:", pump.send_command("Get Serial Number"))
pump.close()
```
